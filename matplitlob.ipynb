```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[10,20,10,20,10]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'b--')   #r for red, b for blue   #'b*--'

plt.show()
```

```python
plt.plot(x,y,'bs--')
plt.show()
```

```python
plt.plot(x,y,'g^--')
plt.show()
```

```python
plt.plot(x,y,'g^--',linewidth=2.0)  #g for green
plt.show()
```

```python
plt.plot(x,y,'g^--',marker='.')
plt.show()
```

```python
t=np.arange(1.,5.,0.2)

#red dashes,blue sqaures,green tables
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g*')
plt.show()
```

```python
x=[1,2,3,4]
y=[20,20,20,20]
plt.plot(x,y,color='green',marker="^",linestyle='dashed',mec='r',mfc='b',linewidth=3,markersize=20)
plt.show()
```
